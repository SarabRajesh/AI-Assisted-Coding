Prompt:- Bubble Sort for Ranking Exam Scores
Write a Python program that implements Bubble Sort to sort a list of student exam scores in ascending order.

Requirements:
1) Add inline comments explaining comparisons, swaps, passes/iterations.
2) Add early termination using a swapped flag (stop if already sorted).
3) Print sample input list and sorted output list clearly.
4) Provide a short time complexity explanation in comments:
   - best case
   - average case
   - worst case

Use this sample list:
scores = [56, 78, 45, 90, 67, 45, 88]

Return:
- Full Python code (single file)
- 5-8 lines explanation of how it works (human, simple)
- Expected output shown exactly as it would appear in terminal

```python
def bubble_sort(scores):
    """
    Bubble Sort implementation with early termination optimization.
    
    Time Complexity:
    - Best case: O(n) - when list is already sorted (swapped flag prevents extra passes)
    - Average case: O(n²) - random order requires multiple passes
    - Worst case: O(n²) - when list is sorted in reverse order
    """
    n = len(scores)
    
    # Outer loop for each pass through the list
    for i in range(n):
        swapped = False  # Flag to detect if any swaps occurred in this pass
        
        # Inner loop for comparing adjacent elements
        for j in range(0, n - i - 1):
            # Compare adjacent scores
            if scores[j] > scores[j + 1]:
                # Swap if left score is greater than right score
                scores[j], scores[j + 1] = scores[j + 1], scores[j]
                swapped = True  # Mark that a swap occurred
        
        # Early termination: if no swaps happened, list is already sorted
        if not swapped:
            break
    
    return scores


# Sample input list
scores = [56, 78, 45, 90, 67, 45, 88]

print("Original Exam Scores:")
print(scores)
print()

# Sort the scores
sorted_scores = bubble_sort(scores.copy())

print("Sorted Exam Scores (Ascending Order):")
print(sorted_scores)
```

Task 2: Improving Sorting for Nearly Sorted Attendance Records

Prompt:- Scenario: Roll numbers are nearly sorted with only a few late updates.

1) First write Bubble Sort (with early termination).
2) Then suggest a better algorithm for nearly sorted data and implement Insertion Sort.
3) Explain clearly (in simple words) why Insertion Sort performs better on nearly sorted data.
4) Compare behavior conceptually on nearly sorted input (number of swaps/shifts idea).

Use this nearly sorted list:
rolls = [101, 102, 103, 105, 104, 106, 107, 109, 108]

Print:
- original list
- bubble sorted output
- insertion sorted output

Return:
- Full code for both algorithms in one Python file
- Short comparison (best/avg/worst complexity) and when to use which
- Expected terminal output

```python
def bubble_sort_optimized(rolls):
    """
    Bubble Sort with early termination.
    
    Time Complexity:
    - Best case: O(n) - nearly sorted data (few passes needed)
    - Average case: O(n²)
    - Worst case: O(n²)
    """
    n = len(rolls)
    
    for i in range(n):
        swapped = False
        
        # Compare adjacent elements and swap if needed
        for j in range(0, n - i - 1):
            if rolls[j] > rolls[j + 1]:
                rolls[j], rolls[j + 1] = rolls[j + 1], rolls[j]
                swapped = True
        
        # Early termination if no swaps occurred
        if not swapped:
            break
    
    return rolls


def insertion_sort(rolls):
    """
    Insertion Sort implementation.
    
    Time Complexity:
    - Best case: O(n) - nearly sorted data (minimal shifts)
    - Average case: O(n²)
    - Worst case: O(n²)
    
    Why better for nearly sorted data: Each element is inserted into its correct position
    in the already-sorted portion. Nearly sorted data requires fewer shifts.
    """
    n = len(rolls)
    
    # Start from second element (first element is already sorted)
    for i in range(1, n):
        key = rolls[i]  # Element to be inserted
        j = i - 1
        
        # Shift elements greater than key one position to the right
        while j >= 0 and rolls[j] > key:
            rolls[j + 1] = rolls[j]
            j -= 1
        
        # Insert key at correct position
        rolls[j + 1] = key
    
    return rolls


# Nearly sorted data
rolls = [101, 102, 103, 105, 104, 106, 107, 109, 108]

print("Original Roll Numbers (Nearly Sorted):")
print(rolls)
print()

# Bubble Sort
bubble_result = bubble_sort_optimized(rolls.copy())
print("After Bubble Sort (Optimized):")
print(bubble_result)
print()

# Insertion Sort
insertion_result = insertion_sort(rolls.copy())
print("After Insertion Sort:")
print(insertion_result)
print()

print("=" * 60)
print("COMPARISON & ANALYSIS")
print("=" * 60)
print("""
WHY INSERTION SORT IS BETTER FOR NEARLY SORTED DATA:

1. Bubble Sort: Makes multiple passes through the entire list,
   repeatedly comparing adjacent elements even if they're already sorted.
   
2. Insertion Sort: Processes elements one by one, inserting each into
   its correct position in the sorted portion. For nearly sorted data,
   elements need minimal shifts, reducing operations significantly.

3. For nearly sorted data like [101, 102, 103, 105, 104, 106, 107, 109, 108]:
   - Bubble Sort: Still performs O(n) passes (iteration overhead)
   - Insertion Sort: Performs O(n) shifts (minimal due to near-sorted state)

BEST ALGORITHM FOR EACH SCENARIO:
- Nearly Sorted Data → Insertion Sort (O(n) best case)
- Random Data → Quick Sort or Merge Sort (O(n log n) average)
- Small Datasets → Insertion Sort (low overhead)
- Large Datasets → Merge Sort or Heap Sort (guaranteed O(n log n))
""")
```

Prompt:- Write a Python program that implements:
1) Linear Search for an UNSORTED list of student roll numbers

unsorted_rolls = [105, 101, 109, 103, 102, 108]

```python
def linear_search(rolls, target):
    """
    Linear Search - searches through each element sequentially.
    
    Parameters:
    - rolls: list of roll numbers (can be unsorted)
    - target: roll number to find
    
    Returns:
    - index of target if found, -1 otherwise
    
    Time Complexity:
    - Best case: O(1) - target is at first position
    - Average case: O(n) - target is somewhere in middle
    - Worst case: O(n) - target is at end or not present
    
    Works on: UNSORTED and SORTED lists
    """
    for i in range(len(rolls)):
        if rolls[i] == target:
            return i  # Element found at index i
    return -1  # Element not found


def binary_search(rolls, target):
    """
    Binary Search - divides sorted list in half repeatedly.
    
    Parameters:
    - rolls: list of roll numbers (MUST BE SORTED)
    - target: roll number to find
    
    Returns:
    - index of target if found, -1 otherwise
    
    Time Complexity:
    - Best case: O(1) - target is at middle position
    - Average case: O(log n) - eliminates half elements each iteration
    - Worst case: O(log n) - target at end or not present
    
    IMPORTANT: Only works on SORTED lists!
    """
    left = 0
    right = len(rolls) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if rolls[mid] == target:
            return mid  # Element found
        elif rolls[mid] < target:
            left = mid + 1  # Search right half
        else:
            right = mid - 1  # Search left half
    
    return -1  # Element not found


# Sample data
unsorted_rolls = [105, 101, 109, 103, 102, 108]
sorted_rolls = sorted(unsorted_rolls)
target = 103

print("=" * 70)
print("LINEAR SEARCH vs BINARY SEARCH")
print("=" * 70)
print()

print("Unsorted Roll Numbers:")
print(unsorted_rolls)
linear_index = linear_search(unsorted_rolls, target)
print(f"Linear Search for {target}: Found at index {linear_index}")
print()

print("Sorted Roll Numbers:")
print(sorted_rolls)
binary_index = binary_search(sorted_rolls, target)
print(f"Binary Search for {target}: Found at index {binary_index}")
print()

print("=" * 70)
print("EXPLANATION & OBSERVATION")
print("=" * 70)
print("""
LINEAR SEARCH:
- Checks each element one by one from start to end
- Works on both sorted and unsorted lists
- Time: O(n) - slower for large datasets
- Used when: Data is unsorted or list is small

BINARY SEARCH:
- Eliminates half the elements with each comparison
- ONLY works on SORTED lists
- Time: O(log n) - much faster for large datasets
- Used when: Data is sorted and speed is critical

STUDENT OBSERVATION:
For our sample of 6 roll numbers, both found the target quickly.
However, imagine 1,000,000 students:
- Linear Search: ~500,000 comparisons (worst case)
- Binary Search: ~20 comparisons (worst case)

That's why sorting first, then using Binary Search is brilliant for
repeated searches! The sort is done once, then searches are lightning fast.
Binary Search is the clear winner for large sorted datasets.
""")
```

```python
def quick_sort(arr):
    """
    Quick Sort (recursive version).

    Parameters:
    - arr: list of comparable elements

    Returns:
    - new sorted list (ascending order)

    The function works by selecting a pivot element (here the middle
    element), partitioning the list into elements less than, equal to,
    and greater than the pivot, then recursively sorting the 'less'
    and 'greater' sublists.  Recursion occurs because each call of
    quick_sort may again split its input until a base case
    (list of length 0 or 1) is reached.
    """
    if len(arr) <= 1:
        # base case: a list with 0 or 1 item is already sorted
        return arr
    pivot = arr[len(arr) // 2]
    less = [x for x in arr if x < pivot]      # elements smaller than pivot
    equal = [x for x in arr if x == pivot]    # elements equal to pivot
    greater = [x for x in arr if x > pivot]   # elements greater than pivot
    # recursive calls sort the partitions
    return quick_sort(less) + equal + quick_sort(greater)


def merge_sort(arr):
    """
    Merge Sort (recursive version).

    Parameters:
    - arr: list of comparable elements

    Returns:
    - new sorted list (ascending order)

    The algorithm recursively splits the list into halves until
    sublists of length 1 or 0 are obtained (base case).  Then it
    merges the sorted halves back together by repeatedly taking the
    smaller front element of each half.
    """
    if len(arr) <= 1:
        # base case
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])    # sort left half
    right = merge_sort(arr[mid:])   # sort right half

    # merge two sorted lists
    merged = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
    # append any remaining elements
    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged


# test data sets
random_data = [7, 2, 9, 1, 5, 8]
sorted_data = [1, 2, 3, 4, 5, 6]
reverse_data = [6, 5, 4, 3, 2, 1]

print("QUICK SORT & MERGE SORT DEMONSTRATION")
print("=" * 60)

for name, dataset in [("random_data", random_data),
                      ("sorted_data", sorted_data),
                      ("reverse_data", reverse_data)]:
    print(f"Testing with {name}: {dataset}")
    print("Quick Sort output:", quick_sort(dataset))
    print("Merge Sort output:", merge_sort(dataset))
    print()

print("=" * 60)
print("COMPARISON & ANALYSIS")
print("=" * 60)
print("""
COMPLEXITY:
- Quick Sort:
    * Best / Average: O(n log n)
    * Worst: O(n²) (poor pivot choices, e.g. already sorted input)
    * In-place (requires only O(log n) stack space), not stable.
- Merge Sort:
    * Best / Average / Worst: O(n log n) guaranteed.
    * Requires O(n) extra space for merging, stable.

PRACTICAL PREFERENCE:
- Use Quick Sort when working with in‑memory arrays and average‑case
  performance is important; its in‑place nature and good cache
  behaviour make it fast in practice.
- Use Merge Sort when stability is required, when dealing with linked
  lists (no random access), or when you need guaranteed O(n log n)
  performance (e.g. for external sorting).  Merge sort is also a
  good choice if you have plenty of auxiliary memory.

Both algorithms are recursive: each call processes a smaller portion
of the original list until trivial base cases are reached, then
combines the results back up the call stack.
""")
```

Task 5: Optimizing Duplicate Detection Algorithm

Prompt:- Write a Python program to detect duplicates in a list of user IDs.

Requirements:
1) Write brute-force duplicate detection using nested loops (O(n^2)).
2) Analyze and print its time complexity in simple words.
3) Suggest an optimized approach using a set (or dictionary) and implement it (O(n)).
4) Print whether duplicates exist and show one duplicate if found (for both methods).
5) Add a conceptual comparison for large inputs (why O(n) is faster than O(n^2)).

Use sample data:
user_ids = [2001, 2005, 2008, 2001, 2010, 2012]

- Explanation of performance improvement

```python
user_ids = [2001, 2005, 2008, 2001, 2010, 2012]

def brute_force_duplicate(ids):
    """
    Brute-force duplicate detection using nested loops.
    Returns (exists: bool, example_duplicate: value or None).
    Time complexity: O(n^2) - compare each element with every other element.
    """
    n = len(ids)
    for i in range(n):
        for j in range(i + 1, n):
            if ids[i] == ids[j]:
                return True, ids[i]
    return False, None


def set_based_duplicate(ids):
    """
    Optimized duplicate detection using a set.
    Returns (exists: bool, example_duplicate: value or None).
    Time complexity: O(n) - single pass, constant-time set checks on average.
    Space complexity: O(n) extra for the set.
    """
    seen = set()
    for x in ids:
        if x in seen:
            return True, x
        seen.add(x)
    return False, None


# Run checks and print results
print("DUPLICATE DETECTION - BRUTE FORCE vs OPTIMIZED")
print("User IDs:", user_ids)
print()

bf_exists, bf_example = brute_force_duplicate(user_ids)
print("Brute-force method (O(n^2)):")
print("Duplicates exist:", bf_exists)
if bf_example is not None:
    print("Example duplicate found:", bf_example)
else:
    print("No duplicate found.")
print("Time complexity (simple): O(n^2) - for each id compare with every other id.")
print()

opt_exists, opt_example = set_based_duplicate(user_ids)
print("Optimized method using set (O(n)):")
print("Duplicates exist:", opt_exists)
if opt_example is not None:
    print("Example duplicate found:", opt_example)
else:
    print("No duplicate found.")
print("Time complexity (simple): O(n) - single pass, checking/adding to a set.")
print()

print("PERFORMANCE COMPARISON (conceptual):")
print("For large inputs, O(n) grows linearly while O(n^2) grows quadratically.")
print("The set-based O(n) approach is much faster on large lists, at the cost of O(n) extra space.")

# Expected terminal output:
# DUPLICATE DETECTION - BRUTE FORCE vs OPTIMIZED
# User IDs: [2001, 2005, 2008, 2001, 2010, 2012]
#
# Brute-force method (O(n^2)):
# Duplicates exist: True
# Example duplicate found: 2001
# Time complexity (simple): O(n^2) - for each id compare with every other id.
#
# Optimized method using set (O(n)):
# Duplicates exist: True
# Example duplicate found: 2001
# Time complexity (simple): O(n) - single pass, checking/adding to a set.
#
# PERFORMANCE COMPARISON (conceptual):
# For large inputs, O(n) grows linearly while O(n^2) grows quadratically.
# The set-based O(n) approach is much faster on large lists, at the cost of O(n) extra space.
```
