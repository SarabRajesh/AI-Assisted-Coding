#1 – Stack Implementation

```python
class Stack:
    """
    A Stack class that follows the Last In, First Out (LIFO) principle.
    Elements are stored internally using a Python list.
    """

    def __init__(self):
        """
        Initializes a new empty Stack.
        """
        self._items = []

    def push(self, item):
        """
        Adds an item to the top of the stack.

        Args:
            item: The item to be added to the stack.
        """
        self._items.append(item)

    def pop(self):
        """
        Removes and returns the item from the top of the stack.

        Returns:
            The item that was at the top of the stack.

        Raises:
            IndexError: If pop() is called on an empty stack.
        """
        if self.is_empty():
            raise IndexError("pop from empty stack")
        return self._items.pop()

    def peek(self):
        """
        Returns the item at the top of the stack without removing it.

        Returns:
            The item at the top of the stack.

        Raises:
            IndexError: If peek() is called on an empty stack.
        """
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self._items[-1]

    def is_empty(self):
        """
        Checks if the stack is empty.

        Returns:
            True if the stack is empty, False otherwise.
        """
        return len(self._items) == 0

    def __len__(self):
        """
        Returns the number of items in the stack.
        """
        return len(self._items)

    def __str__(self):
        """
        Returns a string representation of the stack.
        """
        return f"Stack({list(self._items)})"

# Example Usage:
print("--- Stack Example Usage ---")

# 1. Create a new stack
my_stack = Stack()
print(f"Is stack empty after creation? {my_stack.is_empty()}")
print(f"Stack: {my_stack}, Size: {len(my_stack)}\n")

# 2. Push elements onto the stack
my_stack.push(10)
my_stack.push(20)
my_stack.push(30)
print(f"After pushing 10, 20, 30: {my_stack}")
print(f"Is stack empty? {my_stack.is_empty()}")
print(f"Stack size: {len(my_stack)}\n")

# 3. Peek at the top element
print(f"Top element (peek): {my_stack.peek()}")
print(f"Stack after peek: {my_stack}\n") # Stack should remain unchanged

# 4. Pop elements from the stack
popped_item = my_stack.pop()
print(f"Popped item: {popped_item}")
print(f"Stack after first pop: {my_stack}")
print(f"Stack size: {len(my_stack)}\n")

popped_item = my_stack.pop()
print(f"Popped item: {popped_item}")
print(f"Stack after second pop: {my_stack}")
print(f"Stack size: {len(my_stack)}\n")

# 5. Check if empty and pop/peek from an empty stack
print(f"Is stack empty before final pop? {my_stack.is_empty()}")
popped_item = my_stack.pop()
print(f"Popped item: {popped_item}")
print(f"Stack after third pop: {my_stack}")
print(f"Is stack empty now? {my_stack.is_empty()}")
print(f"Stack size: {len(my_stack)}\n")

# Attempt to pop from an empty stack (should raise IndexError)
try:
    my_stack.pop()
except IndexError as e:
    print(f"Caught expected error when popping from empty stack: {e}")

# Attempt to peek from an empty stack (should raise IndexError)
try:
    my_stack.peek()
except IndexError as e:
    print(f"Caught expected error when peeking from empty stack: {e}")
```

#2 – Queue Implementation

```python
class Queue:
    """
    A Queue class that follows the First In, First Out (FIFO) principle.
    Elements are stored internally using a Python list.
    """

    def __init__(self):
        """
        Initializes a new empty Queue.
        """
        self._items = []

    def enqueue(self, item):
        """
        Adds an item to the rear of the queue.

        Args:
            item: The item to be added to the queue.
        """
        self._items.append(item)

    def dequeue(self):
        """
        Removes and returns the item from the front of the queue.

        Returns:
            The item that was at the front of the queue.

        Raises:
            IndexError: If dequeue() is called on an empty queue.
        """
        if self.is_empty():
            raise IndexError("dequeue from empty queue")
        return self._items.pop(0)

    def peek(self):
        """
        Returns the item at the front of the queue without removing it.

        Returns:
            The item at the front of the queue.

        Raises:
            IndexError: If peek() is called on an empty queue.
        """
        if self.is_empty():
            raise IndexError("peek from empty queue")
        return self._items[0]

    def is_empty(self):
        """
        Checks if the queue is empty.

        Returns:
            True if the queue is empty, False otherwise.
        """
        return len(self._items) == 0

    def size(self):
        """
        Returns the number of items in the queue.
        """
        return len(self._items)

    def __len__(self):
        """
        Returns the number of items in the queue (alias for size()).
        """
        return self.size()

    def __str__(self):
        """
        Returns a string representation of the queue.
        """
        return f"Queue({list(self._items)})"

# Example Usage:
print("--- Queue Example Usage ---")

# 1. Create a new queue
my_queue = Queue()
print(f"Is queue empty after creation? {my_queue.is_empty()}")
print(f"Queue: {my_queue}, Size: {my_queue.size()}\n")

# 2. Enqueue elements onto the queue
my_queue.enqueue('A')
my_queue.enqueue('B')
my_queue.enqueue('C')
print(f"After enqueuing 'A', 'B', 'C': {my_queue}")
print(f"Is queue empty? {my_queue.is_empty()}")
print(f"Queue size: {my_queue.size()}\n")

# 3. Peek at the front element
print(f"Front element (peek): {my_queue.peek()}")
print(f"Queue after peek: {my_queue}\n") # Queue should remain unchanged

# 4. Dequeue elements from the queue
dequeued_item = my_queue.dequeue()
print(f"Dequeued item: {dequeued_item}")
print(f"Queue after first dequeue: {my_queue}")
print(f"Queue size: {my_queue.size()}\n")

dequeued_item = my_queue.dequeue()
print(f"Dequeued item: {dequeued_item}")
print(f"Queue after second dequeue: {my_queue}")
print(f"Queue size: {my_queue.size()}\n")

# 5. Check if empty and dequeue/peek from an empty queue
print(f"Is queue empty before final dequeue? {my_queue.is_empty()}")
dequeued_item = my_queue.dequeue()
print(f"Dequeued item: {dequeued_item}")
print(f"Queue after third dequeue: {my_queue}")
print(f"Is queue empty now? {my_queue.is_empty()}")
print(f"Queue size: {my_queue.size()}\n")

# Attempt to dequeue from an empty queue (should raise IndexError)
try:
    my_queue.dequeue()
except IndexError as e:
    print(f"Caught expected error when dequeuing from empty queue: {e}")

# Attempt to peek from an empty queue (should raise IndexError)
try:
    my_queue.peek()
except IndexError as e:
    print(f"Caught expected error when peeking from empty queue: {e}")
```

#3 – Linked List

```python
class Node:
    """
    Represents a node in a singly linked list.

    Attributes:
        data: The data stored within the node.
        next: A reference to the next node in the list, or None if it's the last node.
    """
    def __init__(self, data):
        """
        Initializes a new Node with the given data.

        Args:
            data: The data to be stored in the node.
        """
        self.data = data
        self.next = None

class LinkedList:
    """
    Represents a singly linked list.

    Attributes:
        head: The head node of the linked list. None if the list is empty.
    """
    def __init__(self):
        """
        Initializes an empty LinkedList.
        """
        self.head = None

    def insert(self, data):
        """
        Inserts a new node with the given data at the end of the linked list.

        Args:
            data: The data to be inserted into the new node.
        """
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def display(self):
        """
        Prints all elements of the linked list in a readable format.
        """
        elements = []
        current = self.head
        while current:
            elements.append(str(current.data))
            current = current.next
        if not elements:
            print("Linked List is Empty")
        else:
            print(" -> ".join(elements))

# Example Usage:
print("--- Singly Linked List Example Usage ---")

# 1. Create a new linked list
my_list = LinkedList()
print("Initial list:")
my_list.display()

# 2. Insert elements
my_list.insert(10)
my_list.insert(20)
my_list.insert(30)
print("List after inserting 10, 20, 30:")
my_list.display()

my_list.insert('Python')
my_list.insert(40.5)
print("List after inserting 'Python', 40.5:")
my_list.display()

# 3. Display an empty list scenario
empty_list = LinkedList()
print("Empty list display test:")
empty_list.display()
```

#4 – Binary Search Tree (BST)


```python
class BST:
    """
    A Binary Search Tree (BST) implementation.
    Maintains the BST property: left subtree values < root < right subtree values.
    """

    class _Node:
        """
        An internal class representing a node in the Binary Search Tree.
        """
        def __init__(self, data):
            """
            Initializes a new Node with the given data.

            Args:
                data: The data to be stored in the node.
            """
            self.data = data
            self.left = None
            self.right = None

    def __init__(self):
        """
        Initializes an empty Binary Search Tree.
        """
        self._root = None

    def insert(self, value):
        """
        Inserts a new value into the BST, maintaining the BST property.
        This method uses recursion.

        Args:
            value: The value to be inserted.
        """
        self._root = self._insert_recursive(self._root, value)

    def _insert_recursive(self, node, value):
        """
        Helper method for recursive insertion.

        Args:
            node: The current node being considered for insertion.
            value: The value to be inserted.

        Returns:
            The node (either the original node or a new node if insertion occurred).
        """
        if node is None:
            return self._Node(value)

        if value < node.data:
            node.left = self._insert_recursive(node.left, value)
        elif value > node.data:
            node.right = self._insert_recursive(node.right, value)
        # If value is equal to node.data, we do nothing (no duplicates allowed or update based on specific requirement)
        return node

    def inorder(self):
        """
        Performs an in-order traversal of the BST and returns a list of its elements.
        In-order traversal visits nodes in the order: Left -> Root -> Right,
        resulting in a sorted list for a BST.

        Returns:
            A list containing the elements of the BST in sorted order.
        """
        elements = []
        self._inorder_recursive(self._root, elements)
        return elements

    def _inorder_recursive(self, node, elements):
        """
        Helper method for recursive in-order traversal.

        Args:
            node: The current node being visited.
            elements: A list to accumulate the visited elements.
        """
        if node:
            self._inorder_recursive(node.left, elements)
            elements.append(node.data)
            self._inorder_recursive(node.right, elements)

# Example Usage:
print("--- Binary Search Tree Example Usage ---")

bst = BST()
print("Initial BST (in-order traversal):", bst.inorder())

print("\nInserting values: 50, 30, 70, 20, 40, 60, 80")
bst.insert(50)
bst.insert(30)
bst.insert(70)
bst.insert(20)
bst.insert(40)
bst.insert(60)
bst.insert(80)

print("BST after insertions (in-order traversal):", bst.inorder())

print("\nInserting another value (25):")
bst.insert(25)
print("BST after inserting 25 (in-order traversal):", bst.inorder())

print("\nInserting a duplicate value (50):")
bst.insert(50) # Assuming duplicates are ignored or handled by not adding
print("BST after inserting duplicate 50 (in-order traversal):", bst.inorder())
```

#5 – Hash Table

```python
class HashTable:
    """
    A HashTable class implementing collision handling using chaining.
    Each bucket in the hash table is a list that can store multiple key-value pairs.
    """

    def __init__(self, size):
        """
        Initializes a new HashTable with a specified size.

        Args:
            size (int): The number of buckets in the hash table.
        """
        self.size = size
        # Initialize the hash table with empty lists (buckets) for chaining
        self.table = [[] for _ in range(self.size)]

    def _hash(self, key):
        """
        A private hash function to compute the index for a given key.

        Args:
            key: The key to be hashed.

        Returns:
            int: The index in the hash table for the given key.
        """
        return hash(key) % self.size

    def insert(self, key, value):
        """
        Inserts a new key-value pair into the hash table, or updates the value
        if the key already exists.

        Args:
            key: The key to be inserted or updated.
            value: The value associated with the key.
        """
        hash_index = self._hash(key)
        bucket = self.table[hash_index]

        # Check if the key already exists in the bucket (collision resolution)
        for i, (k, v) in enumerate(bucket):
            if k == key:
                # Update the value if the key is found
                bucket[i] = (key, value)
                return

        # If key not found, add the new key-value pair to the end of the bucket
        bucket.append((key, value))

    def search(self, key):
        """
        Searches for a key in the hash table and returns its associated value.

        Args:
            key: The key to search for.

        Returns:
            The value associated with the key if found, otherwise None.
        """
        hash_index = self._hash(key)
        bucket = self.table[hash_index]

        # Iterate through the bucket to find the key
        for k, v in bucket:
            if k == key:
                return v  # Key found, return its value
        return None  # Key not found

    def delete(self, key):
        """
        Deletes a key-value pair from the hash table.

        Args:
            key: The key of the item to be deleted.
        """
        hash_index = self._hash(key)
        bucket = self.table[hash_index]

        # Iterate through the bucket to find and remove the key-value pair
        for i, (k, v) in enumerate(bucket):
            if k == key:
                del bucket[i]  # Remove the item from the bucket
                print(f"Key '{key}' deleted.")
                return
        print(f"Key '{key}' not found, cannot delete.")

# Example Usage:
print("--- Hash Table Example Usage ---")

# 1. Create a hash table with 10 buckets
hash_table = HashTable(10)
print(f"Initial Hash Table (size {hash_table.size}): {hash_table.table}\n")

# 2. Insert multiple key-value pairs (including potential collisions)
print("Inserting elements:")
hash_table.insert("apple", 10)
hash_table.insert("banana", 20)
hash_table.insert("cherry", 30)
hash_table.insert("date", 40) # Assume 'date' and 'apple' might hash to same bucket due to small size
hash_table.insert("elderberry", 50)
hash_table.insert("fig", 60)
hash_table.insert("grape", 70)

# Demonstrate updating a value
hash_table.insert("apple", 15) # Update 'apple's value

print("Hash Table after insertions:")
for i, bucket in enumerate(hash_table.table):
    if bucket:
        print(f"Bucket {i}: {bucket}")
print("\n")

# 3. Search for keys
print("Searching for keys:")
print(f"Value for 'banana': {hash_table.search('banana')}")  # Should be 20
print(f"Value for 'apple': {hash_table.search('apple')}")    # Should be 15 (updated value)
print(f"Value for 'mango': {hash_table.search('mango')}")    # Should be None (not found)
print(f"Value for 'date': {hash_table.search('date')}")      # Should be 40
print("\n")

# 4. Delete a key
print("Deleting keys:")
hash_table.delete("cherry")  # Delete an existing key
hash_table.delete("mango")   # Attempt to delete a non-existent key

print("Hash Table after deletions:")
for i, bucket in enumerate(hash_table.table):
    if bucket:
        print(f"Bucket {i}: {bucket}")
print("\n")

# Verify deletion and search again
print("Verification after deletion:")
print(f"Value for 'cherry': {hash_table.search('cherry')}") # Should be None
print(f"Value for 'banana': {hash_table.search('banana')}") # Should still be 20
```

#6 – Graph Representation

```python
class Graph:
    """
    A Graph class implemented using an adjacency list representation.
    Each key in the dictionary represents a vertex, and its value is a list
    of vertices connected to it.
    """

    def __init__(self):
        """
        Initializes an empty graph with no vertices or edges.
        """
        self._adjacency_list = {}

    def add_vertex(self, vertex):
        """
        Adds a new vertex to the graph.

        Args:
            vertex: The vertex to be added.
        """
        if vertex not in self._adjacency_list:
            self._adjacency_list[vertex] = []
            print(f"Vertex '{vertex}' added.")
        else:
            print(f"Vertex '{vertex}' already exists.")

    def add_edge(self, vertex1, vertex2):
        """
        Adds an edge between two vertices in the graph. Assumes an undirected graph.
        If either vertex does not exist, it will be added.

        Args:
            vertex1: The first vertex of the edge.
            vertex2: The second vertex of the edge.
        """
        if vertex1 not in self._adjacency_list:
            self.add_vertex(vertex1)
        if vertex2 not in self._adjacency_list:
            self.add_vertex(vertex2)

        if vertex2 not in self._adjacency_list[vertex1]:
            self._adjacency_list[vertex1].append(vertex2)
            print(f"Edge between '{vertex1}' and '{vertex2}' added.")
        else:
            print(f"Edge between '{vertex1}' and '{vertex2}' already exists.")

        if vertex1 not in self._adjacency_list[vertex2]:
            self._adjacency_list[vertex2].append(vertex1)



    def display(self):
        """
        Prints the graph's adjacency list in a readable format.
        """
        print("--- Graph Adjacency List ---")
        if not self._adjacency_list:
            print("Graph is empty.")
            return

        for vertex, neighbors in self._adjacency_list.items():
            print(f"{vertex}: {', '.join(map(str, neighbors))}")
        print("----------------------------")

# Example Usage:
print("--- Graph Example Usage ---")

# 1. Create a new graph
my_graph = Graph()
my_graph.display()
print("\n")

# 2. Add vertices
my_graph.add_vertex('A')
my_graph.add_vertex('B')
my_graph.add_vertex('C')
my_graph.add_vertex('D')
my_graph.add_vertex('C') # Attempt to add duplicate vertex
print("\n")

# 3. Add edges
my_graph.add_edge('A', 'B')
my_graph.add_edge('A', 'C')
my_graph.add_edge('B', 'D')
my_graph.add_edge('C', 'D')
my_graph.add_edge('E', 'A') # Add edge with non-existing vertex 'E'
my_graph.add_edge('A', 'B') # Attempt to add duplicate edge
print("\n")

# 4. Display the graph
my_graph.display()

# 5. Add a new isolated vertex and display
print("\nAdding isolated vertex 'F':")
my_graph.add_vertex('F')
my_graph.display()
```

#7 – Priority Queue


```python
import heapq

class PriorityQueue:
    """
    A Priority Queue class implemented using Python's heapq module.
    It simulates a max-priority queue, where items with higher numerical priority values
    are dequeued first.
    """

    def __init__(self):
        """
        Initializes an empty PriorityQueue.
        The internal heap stores elements as tuples: (negated_priority, counter, item).
        The counter is used to maintain FIFO order for items with the same priority.
        """
        self._heap = []
        self._counter = 0 # Unique sequence count to ensure FIFO for equal priorities

    def enqueue(self, item, priority):
        """
        Adds an item to the priority queue with a given priority.
        Higher priority values mean higher precedence (max-priority queue).

        Args:
            item: The item to be added to the queue.
            priority (int or float): The priority of the item.
        """
        # We store (-priority, counter, item) to ensure:
        # 1. heapq works as a max-heap (by negating priority).
        # 2. FIFO order for items with the same priority (using counter as a tie-breaker).
        heapq.heappush(self._heap, (-priority, self._counter, item))
        self._counter += 1

    def dequeue(self):
        """
        Removes and returns the item with the highest priority from the queue.

        Returns:
            The item that was in the queue with the highest priority.

        Raises:
            IndexError: If dequeue() is called on an empty queue.
        """
        if self.is_empty():
            raise IndexError("dequeue from empty priority queue")
        # heappop returns the tuple (-priority, counter, item), we only need the item
        neg_priority, count, item = heapq.heappop(self._heap)
        return item

    def peek(self):
        """
        Returns the item with the highest priority without removing it.

        Returns:
            The item that is currently in the queue with the highest priority.

        Raises:
            IndexError: If peek() is called on an empty queue.
        """
        if self.is_empty():
            raise IndexError("peek from empty priority queue")
        # heap[0] gives the smallest element in the heap (which corresponds to highest priority when negated)
        neg_priority, count, item = self._heap[0]
        return item

    def is_empty(self):
        """
        Checks if the priority queue is empty.

        Returns:
            True if the queue is empty, False otherwise.
        """
        return len(self._heap) == 0

    def display(self):
        """
        Displays all elements currently in the queue, along with their original priorities.
        The elements are sorted for display purposes by priority (descending)
        and then by insertion order (ascending) for items with equal priority.
        Note that the internal heap structure does not maintain this full sorted order,
        only the heap property.
        """
        print("--- Priority Queue Elements ---")
        if self.is_empty():
            print("Priority Queue is empty.")
            return

        # Create a list for displaying elements, reconstructing original priorities
        display_list = []
        for neg_prio, count, item in self._heap:
            display_list.append((abs(neg_prio), count, item))

        # Sort for display: highest priority first, then by insertion order for ties
        display_list.sort(key=lambda x: (-x[0], x[1]))

        print("Elements (Priority, Item):")
        for priority, count, item in display_list:
            print(f"  (Priority: {priority}, Item: {item})")
        print("----------------------------")

# Example Usage:
print("--- Priority Queue Example Usage ---")

# 1. Create a new priority queue
pq = PriorityQueue()
print(f"Is PQ empty after creation? {pq.is_empty()}")
pq.display()
print("\n")

# 2. Enqueue elements with different priorities
print("Enqueuing elements:")
pq.enqueue('Task A', 3)
pq.enqueue('Task B', 1)
pq.enqueue('Task C', 5)
pq.enqueue('Task D', 2)
pq.enqueue('Task E', 5) # Same priority as Task C
pq.enqueue('Task F', 4)
pq.display()
print("\n")

# 3. Peek at the highest priority element
print(f"Highest priority element (peek): {pq.peek()}") # Should be 'Task C' or 'Task E'
pq.display() # Queue should remain unchanged
print("\n")

# 4. Dequeue elements
print("Dequeuing elements:")
dequeued_item = pq.dequeue()
print(f"Dequeued item: {dequeued_item}") # Should be 'Task C' (priority 5, first inserted)
pq.display()
print("\n")

dequeued_item = pq.dequeue()
print(f"Dequeued item: {dequeued_item}") # Should be 'Task E' (priority 5, second inserted)
pq.display()
print("\n")

dequeued_item = pq.dequeue()
print(f"Dequeued item: {dequeued_item}") # Should be 'Task F' (priority 4)
pq.display()
print("\n")

# 5. Check empty and attempt dequeue/peek from empty queue
print(f"Is PQ empty before further dequeues? {pq.is_empty()}")
dequeued_item = pq.dequeue()
print(f"Dequeued item: {dequeued_item}") # Should be 'Task A' (priority 3)
pq.display()
print("\n")

dequeued_item = pq.dequeue()
print(f"Dequeued item: {dequeued_item}") # Should be 'Task D' (priority 2)
pq.display()
print("\n")

dequeued_item = pq.dequeue()
print(f"Dequeued item: {dequeued_item}") # Should be 'Task B' (priority 1)
pq.display()
print("\n")

print(f"Is PQ empty now? {pq.is_empty()}")

# Attempt to dequeue from an empty queue (should raise IndexError)
try:
    pq.dequeue()
except IndexError as e:
    print(f"Caught expected error when dequeuing from empty PQ: {e}")

# Attempt to peek from an empty queue (should raise IndexError)
try:
    pq.peek()
except IndexError as e:
    print(f"Caught expected error when peeking from empty PQ: {e}")
```

#8 – Deque

```python
import collections

class DequeDS:
    """
    A Double-Ended Queue (Deque) implementation using Python's collections.deque.
    Allows for efficient additions and removals from both ends.
    """

    def __init__(self):
        """
        Initializes a new empty DequeDS.
        """
        self._deque = collections.deque()

    def insert_front(self, item):
        """
        Inserts an item at the front of the deque.

        Args:
            item: The item to be inserted.
        """
        self._deque.appendleft(item)

    def insert_rear(self, item):
        """
        Inserts an item at the rear of the deque.

        Args:
            item: The item to be inserted.
        """
        self._deque.append(item)

    def remove_front(self):
        """
        Removes and returns the item from the front of the deque.

        Returns:
            The item that was at the front of the deque.

        Raises:
            IndexError: If remove_front() is called on an empty deque.
        """
        if not self._deque:
            raise IndexError("remove_front from empty deque")
        return self._deque.popleft()

    def remove_rear(self):
        """
        Removes and returns the item from the rear of the deque.

        Returns:
            The item that was at the rear of the deque.

        Raises:
            IndexError: If remove_rear() is called on an empty deque.
        """
        if not self._deque:
            raise IndexError("remove_rear from empty deque")
        return self._deque.pop()

    def display(self):
        """
        Prints all elements of the deque in a readable format.
        """
        if not self._deque:
            print("Deque is Empty")
        else:
            print("Deque: [" + ", ".join(map(str, self._deque)) + "]")

    def __len__(self):
        """
        Returns the number of items in the deque.
        """
        return len(self._deque)

# Example Usage:
print("--- DequeDS Example Usage ---")

# 1. Create a new deque
my_deque = DequeDS()
my_deque.display()
print(f"Initial deque size: {len(my_deque)}\n")

# 2. Insert elements at the rear
my_deque.insert_rear(10)
my_deque.insert_rear(20)
my_deque.insert_rear(30)
print("After inserting 10, 20, 30 at rear:")
my_deque.display()
print(f"Deque size: {len(my_deque)}\n")

# 3. Insert elements at the front
my_deque.insert_front(5)
my_deque.insert_front(0)
print("After inserting 5, 0 at front:")
my_deque.display()
print(f"Deque size: {len(my_deque)}\n")

# 4. Remove elements from the front
removed_front_item = my_deque.remove_front()
print(f"Removed from front: {removed_front_item}")
print("Deque after removing from front:")
my_deque.display()
print(f"Deque size: {len(my_deque)}\n")

# 5. Remove elements from the rear
removed_rear_item = my_deque.remove_rear()
print(f"Removed from rear: {removed_rear_item}")
print("Deque after removing from rear:")
my_deque.display()
print(f"Deque size: {len(my_deque)}\n")

# 6. Empty the deque and demonstrate error handling
print("Emptying the deque:")
while len(my_deque) > 0:
    print(f"Removed: {my_deque.remove_front()}")
my_deque.display()
print(f"Deque size: {len(my_deque)}\n")

# Attempt to remove from an empty deque (should raise IndexError)
try:
    my_deque.remove_front()
except IndexError as e:
    print(f"Caught expected error when removing from empty deque (front): {e}")

try:
    my_deque.remove_rear()
except IndexError as e:
    print(f"Caught expected error when removing from empty deque (rear): {e}")
```

#9 Real-Time Application Challenge – Choose the Right Data Structure


```python
import heapq
from datetime import datetime

class LibrarySystem:
    """A system to manage borrowed books using a priority queue."""

    def __init__(self):
        """Initialize an empty priority queue."""
        self._borrowed_books = []

    def borrow_book(self, book_name, due_date_str):
        """
        Add a borrowed book with its due date.

        :param book_name: Name of the book
        :param due_date_str: Due date in 'YYYY-MM-DD' format
        """
        due_date = datetime.strptime(due_date_str, "%Y-%m-%d")
        heapq.heappush(self._borrowed_books, (due_date, book_name))
        print(f"Book '{book_name}' borrowed, due on {due_date_str}.")

    def return_next_due(self):
        """
        Remove and return the book with the nearest due date.

        :raises IndexError: If no books are borrowed
        """
        if not self._borrowed_books:
            raise IndexError("No borrowed books available.")

        due_date, book_name = heapq.heappop(self._borrowed_books)
        return f"Returned '{book_name}' (Due: {due_date.date()})"

    def display_books(self):
        """
        Display all borrowed books sorted by due date.
        """
        if not self._borrowed_books:
            print("No borrowed books.")
            return

        print("Borrowed Books (sorted by due date):")
        # Create a sorted copy for display without modifying the heap
        for due_date, book_name in sorted(self._borrowed_books):
            print(f"{book_name} - Due: {due_date.date()}")


# Example Usage
if __name__ == "__main__":
    library = LibrarySystem()

    library.borrow_book("Data Structures", "2026-03-01")
    library.borrow_book("Python Programming", "2026-02-25")
    library.borrow_book("Algorithms", "2026-02-28")

    library.display_books()

    print("\nReturning the next due book...")
    print(library.return_next_due())

    print("\nBooks after returning one:")
    library.display_books()

    print("\nReturning another next due book...")
    print(library.return_next_due())

    print("\nBooks after returning another:")
    library.display_books()

    print("\nAttempting to return from an empty library...")
    print(library.return_next_due())

    # Attempt to return from an empty library (should raise IndexError)
    try:
        library.return_next_due()
    except IndexError as e:
        print(f"Caught expected error: {e}")
```

#10: Smart E-Commerce Platform – Data Structure Challenge


```python
import collections

"""
Smart E-Commerce Product Search Engine
Implemented using a Hash Table (dictionary-based chaining concept).
Provides fast lookup using Product ID.
"""

class ProductSearch:
    """A simple product search engine using a hash table."""

    def __init__(self, size=10):
        """Initialize the hash table with empty buckets."""
        self.size = size
        self.table = [[] for _ in range(size)]

    def _hash(self, product_id):
        """
        Generate hash index for a given product ID.

        :param product_id: Unique identifier of product
        :return: Index value
        """
        return hash(product_id) % self.size

    def add_product(self, product_id, product_name):
        """
        Add a new product or update existing product.

        :param product_id: Unique product ID
        :param product_name: Name of the product
        """
        index = self._hash(product_id)
        bucket = self.table[index]

        for i, (pid, _) in enumerate(bucket):
            if pid == product_id:
                bucket[i] = (product_id, product_name)
                print(f"Product {product_id} updated.")
                return

        bucket.append((product_id, product_name))
        print(f"Product {product_id} added.")

    def search_product(self, product_id):
        """
        Search for a product by product ID.

        :param product_id: Product ID to search
        :return: Product name if found, else None
        """
        index = self._hash(product_id)
        bucket = self.table[index]

        for pid, name in bucket:
            if pid == product_id:
                return f"Product Found: {name}"

        return "Product not found."

    def delete_product(self, product_id):
        """
        Delete a product using product ID.

        :param product_id: Product ID to delete
        """
        index = self._hash(product_id)
        bucket = self.table[index]

        for i, (pid, _) in enumerate(bucket):
            if pid == product_id:
                del bucket[i]
                print(f"Product {product_id} deleted.")
                return
        print(f"Product {product_id} not found, cannot delete.")

    def display_all_products(self):
        """
        Displays all products currently in the hash table.
        """
        print("\n--- Current Product Catalog ---")
        is_empty = True
        for i, bucket in enumerate(self.table):
            if bucket:
                is_empty = False
                print(f"Bucket {i}: {bucket}")
        if is_empty:
            print("Catalog is empty.")
        print("-------------------------------")


# Example Usage:
print("--- E-Commerce Product Search Engine Example ---")

# 1. Create a product search engine
search_engine = ProductSearch(size=5) # Small size to demonstrate collisions easily
search_engine.display_all_products()

# 2. Add products
print("\nAdding products:")
search_engine.add_product("P001", "Laptop")
search_engine.add_product("P002", "Mouse")
search_engine.add_product("P003", "Keyboard")
search_engine.add_product("P004", "Monitor")
search_engine.add_product("P005", "Webcam")
search_engine.add_product("P006", "Headphones") # Potential collision with small size
search_engine.display_all_products()

# 3. Update a product
print("\nUpdating product P001:")
search_engine.add_product("P001", "Gaming Laptop")
search_engine.display_all_products()

# 4. Search for products
print("\nSearching for products:")
print(search_engine.search_product("P003"))
print(search_engine.search_product("P001"))
print(search_engine.search_product("P999")) # Non-existent product

# 5. Delete products
print("\nDeleting products:")
search_engine.delete_product("P002")
search_engine.delete_product("P007") # Non-existent product
search_engine.display_all_products()

print("\nFinal search for deleted product:")
print(search_engine.search_product("P002"))
```
